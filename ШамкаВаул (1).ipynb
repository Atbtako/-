### Отношение региона к уровню счастья в нем и к здоровому образу жизни

```python
df[['region', 'happiness_score', 'healthy_life_expectancy']].groupby('region').sum()
```

### Вывод:



В Sub-Saharan Africa самые счастливые люди, а в Africa наоборот - самые несчастные.
А так же, в странах, в которых люди больше занимаются спортом люде счастливее.

```python
df2 = df.groupby('region')['happiness_score'].mean().reset_index()
```

```python
plt.figure(figsize=(35, 8))
plt.bar(x=df2['region'], height=df2['happiness_score'])
plt.title("Отноешние регионов и уровня счастья")
```

### Отношение Воли слова к отношению к корупции.

```python
df = df.fillna(1)

numeric_columns = df.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Отношения')
plt.show()
```

### Вывод:


В странах, в которых социальная помощь выше, люди жизнерадостнее. Так же работает и со здоровым образом жизни.

```python
df1 = df[['gdp_per_capita', 'generosity']].head(30)
df1
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_per_capita', y='generosity', data=df1)
plt.title('ВВП на Щедрость')
plt.xlabel('ВВП на душу')
plt.ylabel('Щедрость')
plt.show()
```

### Вывод:

В среднем при большем ВВП на душу населения, Щедрость выше.
