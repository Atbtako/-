## Форматирование датасета для обучения модели данных

```python
column = ['Temperature (K)', 'Luminosity (L/Lo)', 'Radius (R/Ro)', 'Absolute magnitude (Mv)', 'Star type']
star_category_cols = pd.get_dummies(star["Star category"], dtype=int)
X = pd.concat([star[column], star_category_cols], axis = 1)
X.shape
```

```python
X.head()
```

```python
y = star['Spectral Class']
```

```python
star['Spectral Class'].value_counts()
```

### KNeighborsClassifier (Классификация ближайших соседей)

## Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
%%time

gnb = GaussianNB()
gnb.fit(X, y)
```

```python
gnb_y_train_predict = gnb.predict(X)
gnb_y_test_predict = gnb.predict(X)

print('train accuracy: ', accuracy_score(y, gnb_y_train_predict, normalize=True))
print('test accuracy: ', accuracy_score(y, gnb_y_test_predict, normalize=True))
```
